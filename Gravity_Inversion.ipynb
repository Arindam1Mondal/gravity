```python
#******INVERSION************

import numpy as np
from scipy.optimize import least_squares

# Define the model function to compute gravity anomaly
def gravity_model(params, latitude, longitude):
    # params: parameters to be estimated (e.g., depth of the basement)
    # latitude, longitude: coordinates of observation points
    depth = params[0]
    # Assuming a simple model where gravity anomaly is inversely proportional to depth
    #gravity_anomaly = -depth / (latitude**2 + longitude**2 + 1)
    gravity_anomaly=(depth)**(latitude+longitude)
    return gravity_anomaly

# Define the objective function to be minimized (least squares misfit)
def objective_function(params, latitude, longitude, observed_gravity):
    predicted_gravity = gravity_model(params, latitude, longitude)
    misfit = predicted_gravity - observed_gravity
   # print(misfit)
    return misfit


# Example dataset (replace with your own data)
data=np.loadtxt('PG_Data.txt')
latitude = data[:,0]  # Latitude coordinates
longitude = data[:,1]  # Longitude coordinates
observed_gravity =data[:,3]  # Observed gravity anomaly

# Initial guess for the depth of the basement
initial_guess =np.arange(1,81,0.1)



# Perform least squares inversion
result = least_squares(objective_function, initial_guess, args=(latitude, longitude, observed_gravity))

# Extract the estimated parameters
estimated_depth = result.x[0]

print("Estimated depth of the basement:", estimated_depth)
```

```python
plt.scatter(initial_guess,estimated_depth)
plt.grid()
```
